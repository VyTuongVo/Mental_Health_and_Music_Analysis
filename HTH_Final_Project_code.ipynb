```python
#What is the percentage of people who listen to music while working compared to those who don't?


i = 0 
count_yes = 0 
count_no = 0

for i in range(len(data['While working'])):
    if data['While working'][i] == 0:
        count_no = count_no + 1
    elif data['While working'][i] == 1:
        count_yes = count_yes + 1


        
total_count = [count_yes, count_no]
print(total_count)

percentages = [count_yes / sum(total_count) * 100, count_no / sum(total_count) * 100]


values = ["Listen to Music", " Don't listen to music"]


plt.figure(figsize=(15, 8))
plt.pie(total_count, labels=values, autopct='%1.1f%%', textprops={'fontsize': 25})
plt.show() 


```

From the Pie graph, we can see that 79% of participants report listening to music while working. On the other hand, only 21% report that they don't.

```python
#Generally, state the percentage of participants who feel "improved" and "no effect", and "Worsen"


i = 0 
count_improve = 0 
count_no_effect = 0
count_worsen = 0 


for i in range(len(data["Music effects"])):
    if data["Music effects"][i] == "Improve":
        count_improve = count_improve + 1
    elif data["Music effects"][i] == "No effect":
        count_no_effect = count_no_effect + 1
    elif data["Music effects"][i] == "Worsen":
        count_worsen = count_worsen + 1


        
total_count = [count_improve, count_no_effect, count_worsen]
print(total_count)

percentages = [count_improve / sum(total_count) * 100, count_no_effect / sum(total_count) * 100, count_worsen/sum(total_count) * 100 ]


values = ["Improvement", "No Effect", " Worsen"]


plt.figure(figsize=(15, 8))
plt.pie(total_count, labels=values, autopct='%1.1f%%', textprops={'fontsize': 25})
plt.show() 


```
